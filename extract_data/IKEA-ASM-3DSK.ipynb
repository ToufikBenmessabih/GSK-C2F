```python
import pandas as pd

# select joints

Joint = ['nose', 'left_eye_inner', 'left_eye', 'left_eye_outer', 'right_eye_inner', 'right_eye', 'right_eye_outer',
               'left_ear', 'right_ear', 'mouth_left', 'mouth_right', 'left_shoulder', 'right_shoulder', 'left_elbow',
               'right_elbow', 'left_wrist', 'right_wrist', 'left_pinky', 'right_pinky', 'left_index', 'right_index',
               'left_thumb', 'right_thumb', 'left_hip', 'right_hip', 'left_knee', 'right_knee', 'left_ankle',
               'right_ankle', 'left_heel', 'right_heel', 'left_foot_index', 'right_foot_index']



joints = []

for joint in Joint:
    joints.append(joint + '_X')
    joints.append(joint + '_Y')
    joints.append(joint + '_Z')
```

```python
import os
import pandas as pd

# Replace with the actual path to your root directory
root_dir = r'D:\Thèse_Toufik\Datasets\IKEA ASM\ikea_asm_dataset_RGB_top\ANU_ikea_dataset_video\MediaPipe\IKEA_ASM'

# Create a dictionary to store DataFrames
dataframes_dict = {}

# Traverse the directory structure
for subdir, _, files in os.walk(root_dir):
    for file in files:
        if file.endswith("3D.csv"):
            # Get the full path to the CSV file
            file_path = os.path.join(subdir, file)

            # Split the path
            parts = file_path.split(os.sep)

            parent_folder_name = parts[-4]
            grandparent_folder_name = parts[-5]
            
            
            # Create a name combining both the grandparent and parent folder names
            combined_name = f"{grandparent_folder_name}:{parent_folder_name}"
            
            # Read the CSV file into a list of lists (skip the first line and remove the first column)
            data = []
            with open(file_path, 'r') as f:
                line_number = 0
                for line in f:
                    line_number += 1
                    if line_number >= 2:  # Skip the first line
                        values = line.strip().split(',')[1:]  # Skip the first column
                        data.append(values)

            # Convert the list of lists to a DataFrame
            skeleton = pd.DataFrame(data)

            # Optionally, you can set column names (assuming `joints` is predefined):
            skeleton.columns = joints

            # Store the DataFrame in the dictionary using the combined name as the key
            dataframes_dict[combined_name] = skeleton
            print(f"Loaded: {combined_name}")

# Now, dataframes_dict contains DataFrames for each CSV file, keyed by their grandparent and parent folder names
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler
scaler = MinMaxScaler()

# Iterate through each DataFrame in the dictionary and normalize it
normalized_dataframes_dict = {}
for key, df in df_dict.items():
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        
    # Select only numeric columns for normalization
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    
    # Normalize the numeric columns using MinMaxScaler
    df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

    # Store the normalized DataFrame in the new dictionary
    normalized_dataframes_dict[key] = df
    print(key)
```

# Save data into npy files

```python
import numpy as np

# Create a directory to store the npy files (optional)
output_directory = 'npy_IKEA_3D_ESK_25fps'

# Create the directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

print('30 fps 3DSkeleton IKEA-ASM features: ')
for key, df in normalized_dataframes_dict.items():
    
    # Replace any ':' in the combined name with '__'
    key = key.replace(':', '__')

    # Save the normalized DataFrame as a .npy file
    output_filepath = os.path.join(output_directory, f"{key}.npy")
    np.save(output_filepath, df.to_numpy())
    
    d=np.load(output_filepath)
    print(key, d.shape)
```
